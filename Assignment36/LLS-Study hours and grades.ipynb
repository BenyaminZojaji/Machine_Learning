### Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

### making dataset for Study-hours and grades

```python
# x -> study hours 
# y -> grade
X = np.random.uniform(0, 90, 250)
Y = abs(X + np.random.uniform(-10, 10, 250) * np.random.uniform(1, 3, 250))

plt.scatter(X, Y, marker='.')
```

#### random line fit

```python
y_pred = X * 1.5
plt.scatter(X, Y, label='original data')
plt.plot(X, y_pred, c='hotpink', lw=2, label='random fitted line')
plt.legend(loc='best')
plt.show()
```

### LLS(linear least-squares) regression

```python
# slope = (X.T X)^-1 * X.T Y
X = X.reshape((250, 1))
Y = Y.reshape((250, 1))
slope = np.matmul(np.linalg.inv(np.matmul(X.T, X)), np.matmul(X.T, Y))
slope = slope[0,0]
slope
```

```python
Y_pred = X * slope
plt.scatter(X, Y, label='original data')
plt.plot(X, Y_pred, c='red', lw=2, label='fitted line')
plt.legend(loc='best')
plt.show()
```

### LLS using scipy

```python
res = stats.linregress(X, Y)
plt.plot(X, Y, 'o', label='original data')
plt.plot(X, res.intercept + res.slope*X, 'r', label='fitted line-scipy', lw=2)
plt.legend(loc='best')
plt.show()
```

### plot both fitted lines in one figure

```python
plt.plot(X, Y, 'o', label='original data')
plt.plot(X, Y_pred, c='red', lw=2, label='fitted line')
plt.plot(X, res.intercept + res.slope*X, 'orange', label='fitted line-scipy', lw=2)
plt.legend(loc='best')
plt.show()
```
