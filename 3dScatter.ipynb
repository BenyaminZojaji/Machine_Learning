```python
import random
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

```python
# Melon width = 8
# Melon length = 14
# Melon weight = 800
# Balloon width = 16
# Balloon length = 18
# Balloon weight = 5

N = 1000
std = 3
std_weight_melon = 100 
std_weight_baloon = 5

Melons_width = np.random.normal(8, std, N)
Melons_length = np.random.normal(14, std, N)
Melons_weight = np.random.normal(800, std_weight_melon, N)

Balloons_width = np.random.normal(16, std, N)
Balloons_length = np.random.normal(18, std, N)
Balloons_weight = np.random.normal(5, std_weight_baloon, N)

width = np.concatenate((Melons_width, Balloons_width))
length = np.concatenate((Melons_length, Balloons_length))
weight = np.concatenate((Melons_weight, Balloons_weight))

X = np.array([width, length, weight]).T

Melons_label = np.zeros(N, dtype='int')
Balloons_label = np.zeros(N, dtype='int')

Y = np.concatenate((Melons_label, Balloons_label))
```

```python
ax = plt.axes(projection='3d')
ax.scatter(X[0:N, 0], X[0:N, 1], X[0:N, 2], c='red', label='Melon')
ax.scatter(X[N: , 0], X[N: , 1], X[N: , 2], marker= '^', c='orange', label='Baloon')

ax.set_xlabel('Width')
ax.set_ylabel('Length')
ax.set_zlabel('Weight')

ax.set_title('fig-1')

ax.legend(loc="best")

# ax.azim = 40     #  <-x->      -60
# ax.dist = 5     #  distance     10
ax.elev = 20      #  ^-z-v        30

plt.show()
```
